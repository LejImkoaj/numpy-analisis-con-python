### Naranjas

calcular el coeficiente angular y lineal para la recta de las naranjas y para la recta de las toronjas. Utiliza la fórmula de mínimos cuadrados para encontrar cada uno.

```python
y = peso_naranja
x = diametro_naranja
n = np.size(x)
```

```python
a = (n*np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x**2) - np.sum(x)**2)
b = np.mean(y) - a*np.mean(x)

print(f'La pendiente es: {a}\nLa intersección es: {b}')
```

```python
np.linalg.norm(x - y)
```

```python
plt.plot(diametro_naranja, peso_naranja)
plt.plot(x, a*x + b, 'r');
```

### Toronjas

```python
y = peso_toronja
x = diametro_toronja
n = np.size(x)
```

```python
a = (n*np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x**2) - np.sum(x)**2)
b = np.mean(y) - a*np.mean(x)

print(f'La pendiente es: {a}\nLa intersección es: {b}')
```

```python
np.linalg.norm(x - y)
```

```python
plt.plot(diametro_toronja, peso_toronja)
plt.plot(x, a*x + b, 'r');
```
